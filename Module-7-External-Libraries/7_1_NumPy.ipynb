# 7.1 - NumPy (Ch7)
NumPy (Numerical Python) is a fundamental package for scientific computing and array-oriented programming with Python.

### Why NumPy?
- Enables fast numerical computations using arrays.
- Provides powerful array-oriented programming tools.
- Essential for working with large datasets and matrices.

## Importing NumPy

* We can use NumPy in our programs by importing it.
* It is common to give NumPy the identifier `np` using the `as` keyword.

````python
import numpy as np
````

## NumPy Arrays vs Python Lists
- NumPy arrays are mutable, linear/sequence data structures similar to lists.
- NumPy arrays are more efficient than Python lists.
- They allow element-wise operations and broadcasting.
- They are typed and fixed-size for performance.


````python
my_list = [1, 2, 3]
print("List:", my_list)
````

````python
my_array = np.array([1, 2, 3])
print("NumPy Array:", my_array)
````

- NumPy arrays are heterogeneous, i.e. can contain data of any type.
- They can also be defined using tuples.

````python
my_array2 = np.array(("Hello", 35, 4.0))
print("NumPy Array 2:", my_array2)
````

## Creating and Inspecting Arrays

- Use `np.array()` to convert lists into NumPy arrays.

````python
a = np.array([[1, 2], [3, 4], [5, 6]])
print(a)
````

- You can also create arrays filled with zeros or ones:

````python
zeros = np.zeros((2, 3))
print("Zeros:\n", zeros)
````

````python
ones = np.ones((3, 2))
print("Ones:\n", ones)
````

- You can also create arrays filled with any custom value:

````python
filled = np.full((2, 2), 7)
print("Filled:\n", filled)
````

- We can check properties like shape, dimensions, and data type.

````python
print("Shape:", a.shape)
print("Dimensions:", a.ndim)
print("Data Type:", a.dtype)
````

- We can reshape a NumPy array using the `.reshape()` method.
- Often, we need to convert a row array (`shape = (n, 0)`) into a column array (`shape = (n,1)`)

````python
# create row
row = np.array([1, 3, 5, 7, 9])
print("Row:", row)
print("Row Shape:", row.shape)
````

````python
# reshape to a column
col = row.reshape(-1,1)
print("Column:", col)
print("Column Shape:", col.shape)
````

## Useful NumPy Functions
Here are examples of basic element-wise arithmetic.

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a:", a)
print("b:", b)
````

- Add and subtract the arrays element by element:

````python
print("Addition:", np.add(a, b))
print("Subtraction:", np.subtract(a, b))
````

- Multiply and divide the arrays element by element:

````python
print("Multiplication:", np.multiply(a, b))
print("Division:", np.divide(a, b))
````

- Exponentiate, sum and average an array:

````python
print("Exponential:", np.exp(a))
print("Sum:", np.sum(b))
print("Mean:", np.mean(a))
````

- Find the maximum of the array, and the index of the maximum:

````python
print("Max:", np.max(b))
print("Index of Max:", np.argmax(b))
````

## Indexing and Slicing
You can retrieve individual elements and slices of arrays.

````python
arr = np.array([10, 20, 30, 40, 50])
print("Element at index 2:", arr[2])
print("Elements from index 1 to 3:", arr[1:4])
````

You can also change elements using indexing.

````python
arr[0] = 99
print("Modified Array:", arr)
````

Arrays can be populated using loops, just like lists.

````python
arr = np.zeros(5)
for i in range(5):
    arr[i] = i * 10
print("Array populated with loop:", arr)
````

## Deep vs Shallow Copies
In computer science, **shallow copy** means copying only the reference to an object, not the actual data. Any change made through one reference will affect the other, because both point to the same memory.

**Deep copy**, on the other hand, creates a new object in memory and copies the data over. This allows changes to the original to not affect the copy.

### Why is this important?
- Shallow copies can lead to **unexpected bugs** when modifying an array.
- This often happens in functions, where a change to a 'copied' array accidentally mutates the original.

`np.copy()` provides a deep copy to prevent this problem.
Let's look at an example to see the difference.

````python
original = np.array([1, 2, 3])

# make shallow copy
shallow = original

# make deep copy
deep = np.copy(original)
````

````python
# modify original
original[0] = 99

print("Original:", original)
print("Shallow Copy:", shallow)
print("Deep Copy:", deep)
````
