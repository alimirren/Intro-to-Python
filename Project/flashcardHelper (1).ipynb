# Functions

This block hides all the functions

```python
import random
```

```python
# This is the main dictionary
flashcards = {}
```

```python
def add_word():
  '''Add a single element to the dictionary'''

  word = input("Enter the word/phrase: ")
  definition = input("Enter its definition: ")
  flashcards[word] = definition
```

```python
def add_words():
  '''Add multiple elements to the dictionary'''

  while True:
    add_word()
    choice = ""
    while choice != 'Y' and choice != 'N':
      print("Would you like to add another word? (\033[32mY\033[0m/\033[31mN\033[0m)")
      choice = input().upper()
    if choice == 'Y':
      continue
    else:
      break
```

```python
def display_flashcards():
  '''Will nicely display the dictionary'''

  print("Your flashcards:\n---------------")
  for word, desc in flashcards.items(): # .items() returns dictionary's tuple pairs :P
    print(f"\033[33m{word}\033[0m: {desc}\n---------------")
```

```python
def study_words():
  '''Study words by flipping the flashcards or moving onto the next'''

  # print the word from dictionary, then give a choice to move or flip
  for word, desc in flashcards.items():
    print('"' + word + '"')
    choice = ""
    print("---------------")

    while choice != "flip" and choice != "move" and choice != "stop":
      print("\033[33mFlip/\033[32mMove/\033[31mStop: \033[0m")
      choice = input().lower()

    if choice == "flip":
      print("\n")
      print(desc)
      print("---------------")
    elif choice == "move":
      print("\n")
    else:
      break
```

```python
def play():
  '''Play a minigame where you guess the word by reading its definition'''
  c = ""

  while c != "stop":
    # Here I choose a random key from flashcards
    c_key = random.choice(list(flashcards.keys())) # choose one random key
    c_value = flashcards[c_key] # store its value

    # Here I make a list of 3 choices
    w_keys = list(flashcards.keys()) # make a list of all keys
    w_keys.remove(c_key) # remove the correct key
    w_choices = random.sample(w_keys, 2) # choose 2 random incorrect values
    choices = [c_key] + w_choices # combine all choices together
    random.shuffle(choices) # shuffle choices

    # Now display it to the user
    print("\n\033[34mGuess the word by its definition!\033[0m")
    print(f'"{c_value}"')

    for num, key in enumerate(choices, 1): # go through the enumerated choices list to print them nicely..
      print(f"\033[33m{num}\033[0m. {key}")
      if key == c_key:
        cc = num

    # Decided to add exception handling because if user enters a string then it would result in a ValueError
    while True:
      try:
        uc = int(input("Correct choice?: : "))
      except ValueError:
        print("\033[31mPlease enter (1,2,3)\033[0m")
      else:
        break

    # check if the user's choice is correct
    if uc == cc:
      print("\033[32mCorrect!\033[0m")
    else:
      print("\033[31mWrong!\033[0m")

    # the only stupid way i could figure out to stop the function
    c = input('"Stop" to quit, Enter to continue: ').lower()
```
