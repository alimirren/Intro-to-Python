## `if`...`else`
* Performs different suites, based on whether a condition is `True` or `False`.
* Pseudocode:

> _If student’s grade is greater than or equal to 60  
> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Display 'Passed'  
> Else  
> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Display 'Failed'_

* Correspondong Python code with variable `grade` initalized to `85`

```python
grade = 85
```

```python
if grade >= 60:
  print("Passed")
else:
  print("Failed")
```

* Assign `57` to `grade`, then shows the `if`…`else` statement again to demonstrate that only the `else` suite executes

```python
grade = 57
```

```python
if grade >= 60:
  print("Passed")
else:
  print("Failed")
```

### Conditional Expressions
* Sometimes the suites in an `if`…`else` statement assign different values to a variable, based on a condition

```python
grade = 87
```

```python
if grade >= 60:
  result = "Passed"
else:
  result = "Failed"
```

```python
print(result)
```

* Can write statements like this using a concise conditional expression.
* The parentheses are not required, but they make it clear that the statement assigns the conditional expression’s value to `result`.

```python
result = "Passed" if grade >= 60 else "Failed"
```

```python
result
```

* In interactive mode, you also can evaluate the conditional expression directly.

```python
"Passed" if grade >= 60 else "Failed"
```

### Multiple Statements in a Suite

```python
grade = 49
```

```python
if grade >= 60:
  print("Passed")
else:
  print("Failed")
  print("You must retake the class")
```

* If you do not indent the second `print`, then it’s not in the `else`’s suite.
* In that case the statement always executes, creating strange incorrect output.

### `if`…`elif`…`else` Statement
* Can test for many cases.
* Only the action for the first `True` condition executes.

Let's write a program that takes as input a user's score out of 100, and returns their grade (i.e. `A`, `B`, etc..)

### Pseudocode

> ask the user to input their score
> store the score as an integer

> if score is more than 90, display "A"

> else if score is more than 80, display "B"

> else if score is more than 70, display "C"

> else if score is more than 60, display "D"

> else display "F"


```python
user_score = int(input("Enter your score: "))
score = ""

if user_score >= 90:
  score = "A"
elif user_score >= 80:
  score = "B"
elif user_score >= 70:
  score = "C"
elif user_score >= 60:
  score = "D"
else:
  score = "F"
print("Your score is {}".format(score))
```

### `else` Is Optional
* Handle values that do not satisfy any of the conditions.
* Without an `else`, if no conditions are `True`, the program does not execute any of the statement’s suites.

### Logic Errors
* For a nonfatal logic error, code executes, but produces incorrect results.
* For a fatal logic error in a script, an exception occurs, Python displays a traceback, then the script terminates.
* A fatal error in interactive mode terminates the current snippet, then IPython waits for your next input.

3. Write a program which completes the following tasks. Before you start writing code, plan out your algorithm using pseudocode, or draw a flowchart.  
  * Ask the user to input their `age`.
  * if the `age` is between 0-12, print `"Your ticket price is $5"`
  * if the `age` is between 13-17, print `"Your ticket price is $8"`
  * if the `age` is between 18-64, print `"Your ticket price is $12"`
  * if the `age` is 65+, print `"Your ticket price is $7"`
  * if the `age` is negative, print an error message `"Age cannot be negative"`

Be sure to test your program with various ages.

```python
age = int(input("Please enter your age: "))

if 0 <= age <= 12:
  print("Your ticket price is $5")
elif 13 <= age <= 17:
  print("Your ticket price is $8")
elif 18 <= age <= 64:
  print("Your ticket price is $12")
elif age >= 65:
  print("Your ticket price is $7")
else:
  print("Age cannot be negative.")
```

4. A restaurant has the following menu items:
  1. Burger: $5.99

  2. Pizza: $8.99

  3. Salad: $6.49

In a single code cell, write a program which completes the following tasks:
  * Print out the menu to the user.
  * Ask the user to input a menu `choice` (this should be an integer: 1,2 or 3)
  * Based on `choice`, print out the price of the chosen item
  * if the user enters an invalid choice, print out an error message `"Invalid choice"`

```python
burger = 5.99
pizza = 8.99
salad = 6.49

print("Welcome! We have burgers, pizza and salads.")
choice = int(input("Enter your choice as a number 1, 2 or 3: "))

if choice == 1:
  print("That would be ${}".format(burger))
elif choice == 2:
  print("That would be ${}".format(pizza))
elif choice == 3:
  print("That would be ${}".format(salad))
else:
  print("Invalid choice")
```

5. (**Basic FizzBuzz**) Write a program that does the following:
  * Asks the user to input a `number`.
  * if the `number` is divisible by 3, print `Fizz`
  * if the `number` is divisible by 5, print `Buzz`
  * if the `number` is divisible by 3 **and** 5, print `FizzBuzz`

_(Hint: plan out your code with pseudocode. Order matter!)_

```python
num = int(input("Enter a number: "))
result = ""

if num % 3 == 0:
  result = "Fizz"
if num % 5 == 0:
  result += "Buzz"

print(result)
```
