# Lecture 4 - Getting Input from the User
***
* Can use `int()`, `float()` and `str()` to cast types.
* Built-in **`input` function** requests and obtains user input.

***
## Type Casting

We can cast floats and strings to integers.

```python
int(15.6)
```

Note that `int()` casts by **rounding down**. It can cast strings:

```python
int("15")
```

If the conversion is not obvious, the casting functions will throw an **Exception**

```python
int("CS171")
```

Casting to `float` behaves in a similar way:

```python
float(4)
```

```python
float("4.5")
```

```python
float("6")
```

```python
float(4.5)
```

Casting to a string will convert almost any type in Python to a string:

```python
str(6)
```

```python
str(7.7)
```

```python
str('op')
```

## Getting Input from the User

```python
name = input("What's your name?: ")
```

```python
name
```

```python
type(name)
```

* If you enter quotes, they’re input as part of the string.

```python
name = input("What's your name?: ")
```

```python
name
```

```python
type(name)
```

### Function `input` Always Returns a String

```python
value1 = input("Enter the first number: ")
```

```python
value2 = input("Enter the second number: ")
```

```python
value1 + value2
```

* Python “adds” the _string_ values `'7'` and `'3'`, producing the _string_ `'73'`.
* Known as **string concatenation**.

### Getting an Integer from the User
* If you need an integer, convert the string to an integer using the built-in **`int` function**.

```python
value = int(input("Enter an integer: "))
value
```

```python
value2 = input("Enter an integer: ")
value2 = int(value2)
value
```

```python
value + value2
```

* Can combine `int` and `input` in one statement.

* If the string passed to `int` cannot be converted to an integer, a `ValueError` occurs.

* Function `int` also can convert a floating-point value to an integer.

```python
int(10.5)
```
