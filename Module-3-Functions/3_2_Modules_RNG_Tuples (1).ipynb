## 4.4 Random-Number Generation
* Can introduce the element of chance via the Python Standard Library’s `random` module.

### Rolling a Six-Sided Die
* Product 10 random integers in the range 1–6 to simulate rolling a six-sided die:

```python
import random
```

```python
for roll in range(10):
  random.seed(42)
  print(random.randrange(1,7), end=' ')
```

* `randrange` function generates an integer from the first argument value up to, but not including, the second argument value.
* Different values are displayed if you re-execute the loop.

* Can guarantee **reproducibility** of a random sequence with the `random` module’s `seed` function.  

### Rolling a Six-Sided Die 6,000,000 Times
* If `randrange` truly produces integers at random, every number in its range has an equal probability (or chance or likelihood) of being returned each time we call it.
* Roll a die 6,000,000 times.
* Each die face should occur approximately 1,000,000 times.
* We used Python’s underscore (_) digit separator to make the value 6000000 more readable.

```python
import random

# face frequency counters
frequency1 = 0
frequency2 = 0
frequency3 = 0
frequency4 = 0
frequency5 = 0
frequency6 = 0

# 6,000,000 die rolls
for roll in range(6_000_000):  # note underscore separators
    face = random.randrange(1, 7)

    # increment appropriate face counter
    if face == 1:
        frequency1 += 1
    elif face == 2:
        frequency2 += 1
    elif face == 3:
        frequency3 += 1
    elif face == 4:
        frequency4 += 1
    elif face == 5:
        frequency5 += 1
    elif face == 6:
        frequency6 += 1

print(f'Face{"Frequency":>13}')
print(f'{1:>4}{frequency1:>13}')
print(f'{2:>4}{frequency2:>13}')
print(f'{3:>4}{frequency3:>13}')
print(f'{4:>4}{frequency4:>13}')
print(f'{5:>4}{frequency5:>13}')
print(f'{6:>4}{frequency6:>13}')
```

## Case Study - Dice Rolling in Monopoly

* Requirements Statement:
> You roll two six-sided dice, each with faces containing one, two, three, four, five and six spots, respectively. The sum of the two dice rolls is the player's allowed movement range. However, if they roll a "double" (two of the same value), they get to roll again, and the sum gets added to their first movement range. If they roll a double again, they get to roll again. If they roll three doubles in a row, then they go to Jail, and their movement range is 0.

We implement this function below.

```python
# Roll Dice in Monopoly
import random

"""Simulate the dice roll in Monopoly"""

def display_dice(dice1, dice2, roll):
  """Display dice outcome nicely"""
  print(f"Roll {roll}")
  print(f"Dice1 = {dice1}, Dice2 = {dice2}\n")


def roll_dice():
  """Roll two dice and return the outcome"""
  dice1 = random.randrange(1,7)
  dice2 = random.randrange(1,7)
  return (dice1, dice2)


def monopoly_roll():
  """Monopoly roll with doubles"""
  sum = 0
  doubles = 0

  while doubles < 3:
    dice1, dice2 = roll_dice()
    display_dice(dice1, dice2, doubles + 1)
    sum += dice1 + dice2

    if dice1 != dice2:
      break
    else:
      doubles += 1

    if doubles == 3:
      print("Go to jail!")
      return 0

  print(f"Total sum is {sum}")
  return sum
```

```python
monopoly_roll()
```

## Packing and Unpacking Tuples

* The last example we returned more than one value, by **packing** the values into a **tuple**.
* **Tuples** are _immutable_ types, meaning they cannot be changed once they are defined.
* We can **unpack** tuples by assigning a tuple to a comma separated list of variables


```python
def square_two_nums(num1, num2):
  # pack the tuple
  return num1**2, num2**2

# unpack the tuple
square1, square2 = square_two_nums(5, 8)

print(f"Square1 = {square1}, Square2 = {square2}")
```

* We can check if a value is within an iterator (like a list or tuple) using the `in` operator. This produces a condition, with `True` if the left operand's value is found in the right operand.

```python
def even_dice_roll(roll):
  return True if roll in (2, 4, 6) else False
```

```python
roll = random.randrange(1, 7)

print(f"Roll = {roll}, Even = {even_dice_roll(roll)}")
```

## Immutable vs Mutable Types

* Mutable means the object can be modified/changed. Immutable means the object cannot be modified. Instead, immutable objects are copied and then changed.
* `int, float, str` and `tuple` are all **immutable**
* `list, dict` and `set` are all **mutable**
